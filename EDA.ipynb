```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from read_data import *


def EDA_full(df):
    # 1. Гистограммы концентраций: пары
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))
    for gas, color in zip(['H2S', 'NO2'], ['orange', 'r']):
        if gas in df.columns:
            df[df[gas] > 0][gas].hist(bins=100, ax=axs[0], alpha=0.5, label=gas, color=color)
    axs[0].set_title("H2S и NO2 (>0)")
    axs[0].set_xlabel("Концентрация, мкг/м³")
    axs[0].set_ylabel("Частота")
    axs[0].legend()

    for gas, color in zip(['CO', 'O3'], ['b', 'm']):
        if gas in df.columns:
            df[df[gas] > 0][gas].hist(bins=100, ax=axs[1], alpha=0.5, label=gas, color=color)
    axs[1].set_title("CO и O3 (>0)")
    axs[1].set_xlabel("Концентрация, мкг/м³")
    axs[1].set_ylabel("Частота")
    axs[1].legend()

    if 'SO2' in df.columns:
        df[df['SO2'] > 0]['SO2'].hist(bins=100, ax=axs[2], color='g', alpha=0.7, label='SO2')
        axs[2].set_title("SO2 (>0)")
        axs[2].set_xlabel("Концентрация, мкг/м³")
        axs[2].set_ylabel("Частота")
        axs[2].legend()
    else:
        axs[2].axis('off')
    plt.tight_layout()
    plt.show()

    # 2. Гистограммы сигналов
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))
    for gas, color in zip(['H2Sop1', 'NO2op1'], ['orange', 'r']):
        if gas in df.columns:
            df[gas].hist(bins=100, ax=axs[0], alpha=0.5, label=gas, color=color)
    axs[0].set_title("H2Sop1 и NO2op1")
    axs[0].set_xlabel("Сигнал op1, мВ")
    axs[0].set_ylabel("Частота")
    axs[0].legend()

    for gas, color in zip(['COop1', 'O3op1'], ['b', 'm']):
        if gas in df.columns:
            df[gas].hist(bins=100, ax=axs[1], alpha=0.5, label=gas, color=color)
    axs[1].set_title("COop1 и O3op1")
    axs[1].set_xlabel("Сигнал op1, мВ")
    axs[1].set_ylabel("Частота")
    axs[1].legend()

    if 'SO2op1' in df.columns:
        df['SO2op1'].hist(bins=100, ax=axs[2], color='g', alpha=0.7, label='SO2op1')
        axs[2].set_title("SO2op1")
        axs[2].set_xlabel("Сигнал op1, мВ")
        axs[2].set_ylabel("Частота")
        axs[2].legend()
    else:
        axs[2].axis('off')
    plt.tight_layout()
    plt.show()

    # 3. Гистограммы температур
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))
    for gas, color in zip(['H2St', 'NO2t'], ['orange', 'r']):
        if gas in df.columns:
            df[gas].hist(bins=100, ax=axs[0], alpha=0.5, label=gas, color=color)
    axs[0].set_title("H2St и NO2t")
    axs[0].set_xlabel("Температура, °C")
    axs[0].set_ylabel("Частота")
    axs[0].legend()

    for gas, color in zip(['COt', 'O3t'], ['b', 'm']):
        if gas in df.columns:
            df[gas].hist(bins=100, ax=axs[1], alpha=0.5, label=gas, color=color)
    axs[1].set_title("COt и O3t")
    axs[1].set_xlabel("Температура, °C")
    axs[1].set_ylabel("Частота")
    axs[1].legend()

    if 'SO2t' in df.columns:
        df['SO2t'].hist(bins=100, ax=axs[2], color='g', alpha=0.7, label='SO2t')
        axs[2].set_title("SO2t")
        axs[2].set_xlabel("Температура, °C")
        axs[2].set_ylabel("Частота")
        axs[2].legend()
    else:
        axs[2].axis('off')
    plt.tight_layout()
    plt.show()

    # 4. Boxplot для концентраций
    fig, axs = plt.subplots(1, 3, figsize=(18, 5))
    cols1 = [c for c in ['H2S', 'NO2'] if c in df.columns]
    cols2 = [c for c in ['CO', 'O3'] if c in df.columns]
    cols3 = [c for c in ['SO2'] if c in df.columns]
    sns.boxplot(data=df[cols1].replace(0, np.nan), showfliers=False, ax=axs[0])
    axs[0].set_title('H2S и NO2')
    sns.boxplot(data=df[cols2].replace(0, np.nan), showfliers=False, ax=axs[1])
    axs[1].set_title('CO и O3')
    if cols3:
        sns.boxplot(data=df[cols3].replace(0, np.nan), showfliers=False, ax=axs[2])
        axs[2].set_title('SO2')
    else:
        axs[2].axis('off')
    plt.tight_layout()
    plt.show()

    # 5. Корреляции
    sig_cols = [c for c in ['H2Sop1', 'NO2op1', 'COop1', 'O3op1', 'SO2op1'] if c in df.columns]
    mt_mh_cols = [c for c in ['MT', 'MH'] if c in df.columns]
    if sig_cols and mt_mh_cols:
        corr_df = df[sig_cols + mt_mh_cols].corr().loc[sig_cols, mt_mh_cols]
        plt.figure(figsize=(6, 4))
        sns.heatmap(corr_df, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
        plt.title("Корреляция сигналов с MT, MH")
        plt.show()

    conc_cols = [c for c in ['H2S', 'NO2', 'CO', 'O3', 'SO2'] if c in df.columns]
    if sig_cols and conc_cols:
        corr_df2 = df[sig_cols + conc_cols].corr().loc[sig_cols, conc_cols]
        plt.figure(figsize=(8, 5))
        sns.heatmap(corr_df2, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
        plt.title("Корреляция сигналов с концентрациями")
        plt.show()

    # 6. Статистики
    from IPython.display import display
    with pd.option_context('display.max_columns', 100):
        print('=== Описательные статистики для концентраций ===')
        display(df[conc_cols].describe())
        print('=== Описательные статистики для сигналов (op1) ===')
        display(df[sig_cols].describe())
        print('=== Описательные статистики для температур и влажности ===')
        temp_hum_cols = [c for c in ['COt', 'NO2t', 'SO2t', 'H2St', 'O3t', 'T', 'MT', 'MH'] if c in df.columns]
        display(df[temp_hum_cols].describe())

    # Маски bl/stat всех вариантов
    mask_names = []
    for name in df.columns:
        if 'bl' in name and df[name].dropna().nunique() <= 2:
            mask_names.append(name)
        if 'stat_auto' in name and df[name].dropna().nunique() <= 2:
            mask_names.append(name)

    for mask in mask_names:
        if mask in df.columns:
            print(f"\n=== Статистики для маски {mask}=True ===")
            display(df[df[mask]==True][conc_cols + sig_cols].describe())

    # Пропуски и уникальные значения
    print("\n=== Количество пропусков по столбцам ===")
    print(df.isna().sum())

    key_cols = ['sensor_id'] if 'sensor_id' in df.columns else []
    key_cols += [c for c in mask_names if c in df.columns]
    for col in key_cols:
        print(f"\nУникальные значения для {col}:")
        print(df[col].value_counts(dropna=False))

    # Аномалии
    print("\n=== Аномалии (отрицательные значения, температуры/влажности вне диапазона) ===")
    humidity_cols = [
        'mh', 'g1_mh', 'g2_mh', 'rh_thc',
        'righth', 'lefth', 'rh_rhtester1', 'rh_rhtester2'
    ]
    for gas in conc_cols:
        if gas in df.columns:
            neg = (df[gas] < 0).sum()
            print(f"{gas}: отрицательных значений — {neg}")

    for tcol in temp_hum_cols:
        if tcol in df.columns:
            col_lc = tcol.lower()
            if any(word in col_lc for word in humidity_cols):
                out_range = ((df[tcol] < 0) | (df[tcol] > 100)).sum()
                print(f"{tcol}: значений вне диапазона [0, 100] — {out_range}")
            else:
                out_range = ((df[tcol] < -40) | (df[tcol] > 70)).sum()
                print(f"{tcol}: значений вне диапазона [-40, 70] — {out_range}")

df = load_lab_data(data_dir='data_example/lab_data', gases=None)
df = add_bl_stat_masks(df, gases=['NO2'], std_thr=1.8, window=7, use_isolation=True, contamination=None, verbose=True)
EDA_full(df)
```
